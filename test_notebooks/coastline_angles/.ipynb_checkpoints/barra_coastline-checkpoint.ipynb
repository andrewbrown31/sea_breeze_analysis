```python
size_thresh=4

#Get coastline and labelled landmass masks using skimage
coastline_masks, label_masks = label_and_find_coast(lsm, 
                                                    size_thresh=size_thresh)
label = coastline_masks[0]

lon = lsm.lon.values
lat = lsm.lat.values
xx,yy = np.meshgrid(lon,lat)

xl, yl = np.where(label)

angle_ls = []
for t in tqdm.tqdm(np.arange(len(xl))):
    angles = np.zeros(label.shape) * np.nan
    for i in range(xx.shape[0]):
        for j in range(xx.shape[1]):
            angles[i,j] = np.arctan2(
                xx[i,j] - xx[xl[t],yl[t]],
                yy[i,j] - yy[xl[t],yl[t]])
    
    angles = np.rad2deg(angles)
    angles = (angles-90) % 360

    angle_ls.append(angles)

plt.pcolormesh(xx,yy,
         scipy.stats.circmean(np.stack(angle_ls), axis=0, high=360, low=0),
    cmap="hsv",vmin=0,vmax=360) ; plt.colorbar()
plt.plot(lon[yl],lat[xl],marker="o",ls="none",color="k",ms=1)
```
