```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
from climtas.nci import GadiClient
import metpy

from working.sea_breeze.load_barra import load_barra_wind_data, load_barra_variable, load_barra_static
```

```python
client = GadiClient()
client
```

```python
t1 = "2024-01-24 00:00"
t2 = "2024-02-02 01:00"
lat_slice = slice(-40,-29)
lon_slice = slice(140,156)

wind_ds = load_barra_wind_data(
    ["uas","ua1000","ua100m","ua925","ua1500m","ua850","ua700"],
    ["vas","va1000","va100m","va925","va1500m","va850","va700"],
    ["ws_sfc","ws_1000","ws_100m","ws_925","ws_1500m","ws_850","ws_700"],
    t1,
    t2,
    "AUS-11",
    "1hr",
    lat_slice,
    lon_slice)


huss = load_barra_variable("huss",t1,t2,"AUS-11","1hr",lat_slice,lon_slice)
ts = load_barra_variable("ts",t1,t2,"AUS-11","1hr",lat_slice,lon_slice)
ps = load_barra_variable("ps",t1,t2,"AUS-11","1hr",lat_slice,lon_slice)

orog, lsm = load_barra_static("AUS-11",lon_slice,lat_slice)
lsm = (lsm.sftlf >= 0.5)
```

```python
def plot_sfc_wind(sfc_wind,ax,vmax=None,vmin=None,cmap=None,ws=True):
    if ws:
        sfc_wind.ws_100m.plot(vmax=vmax,vmin=vmin,cmap=cmap)
    sfc_wind[["ua100m","va100m"]].coarsen({"lon":2,"lat":2},boundary="pad").mean().plot.quiver(x="lon",y="lat",u="ua100m",v="va100m")
    ax.coastlines()
    ax.gridlines(draw_labels=True)

def plot_ds(ds,var_name,ax,vmin=None,vmax=None,cmap=None):
    ds[var_name].plot(ax=ax,vmin=vmin,vmax=vmax,cmap=cmap)
```

```python
plt.figure(figsize=[12,6])
ax=plt.axes(projection=ccrs.PlateCarree())
plot_sfc_wind(wind_ds.mean("time"), ax, vmin=0, vmax=10)
plt.gca().plot(148,-38,marker="x",color="tab:orange")
plt.gca().plot(147.8,-37.8,marker="x",color="tab:orange")
plt.plot([146,150],[-38.8,-37.3])
```

```python
xr.where(lsm,ts.ts,np.nan).mean(("lat","lon")).plot()
xr.where(~lsm,ts.ts,np.nan).mean(("lat","lon")).plot()
```

```python
ls_c = xr.where(lsm,ts.ts,np.nan).mean(("lat","lon")) - xr.where(~lsm,ts.ts,np.nan).mean(("lat","lon"))


dc_dt = (metpy.constants.dry_air_gas_constant * np.log( 1000 / 900 ) * ls_c)
P = 2 * (50000+1000)
dv_dt = dc_dt / P
(dv_dt).plot()
```

```python
dv_dt.sel(time=slice("2024-01-23 12","2024-01-24 12:00")).plot()
```

```python
dv_dt_slice = dv_dt.sel(time=slice("2024-01-24 04:00","2024-01-24 07:00"))
seconds = np.arange(0, dv_dt.sel(time=slice("2024-01-24 04:00","2024-01-24 07:00")).time.shape[0] * 60 * 60, 3600)

np.trapz(dv_dt_slice,seconds)
```

```python
#Calculating frontogenesis

u = wind_ds.uas
v = wind_ds.vas
q = metpy.calc.mixing_ratio_from_specific_humidity(huss.huss)
theta_v = metpy.calc.virtual_potential_temperature(ps.ps,ts.ts,q)
theta = metpy.calc.potential_temperature(ps.ps,ts.ts)
f = metpy.calc.frontogenesis(theta,u,v,x_dim=2, y_dim=1)
f = f.metpy.convert_units(metpy.units.units.kelvin / metpy.units.units.kilometer / metpy.units.units.hour)
```

```python
thresh = 0.1 * metpy.units.units.kelvin / metpy.units.units.kilometer / metpy.units.units.hour

plt.figure()
ax=plt.axes(projection=ccrs.PlateCarree())
((f.sel(time=np.in1d(f.time.dt.hour,0))>thresh) & (wind_ds.ws_700<=15)).mean("time").plot(cmap="Blues",vmin=0,vmax=0.4)
#ax.coastlines()

plt.figure()
ax=plt.axes(projection=ccrs.PlateCarree())
((f.sel(time=np.in1d(f.time.dt.hour,6))>thresh) & (wind_ds.ws_700<=15)).mean("time").plot(cmap="Blues",vmin=0,vmax=0.4)
#ax.coastlines()

plt.figure()
ax=plt.axes(projection=ccrs.PlateCarree())
((f.sel(time=np.in1d(f.time.dt.hour,12))>thresh) & (wind_ds.ws_700<=15)).mean("time").plot(cmap="Blues",vmin=0,vmax=0.4)
#ax.coastlines()

plt.figure()
ax=plt.axes(projection=ccrs.PlateCarree())
((f.sel(time=np.in1d(f.time.dt.hour,18))>thresh) & (wind_ds.ws_700<=15)).mean("time").plot(cmap="Blues",vmin=0,vmax=0.4)
#ax.coastlines()
```
