```python
from sea_breeze import load_model_data
import xarray as xr
from dask.distributed import Client
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import hvplot.xarray
import numpy as np
import logging
logging.getLogger("flox").setLevel(logging.WARNING)
```

```python
Client()
```

```python
angle_list = []
models = ["aus2200","barra_c","barra_r","era5"]
for m in models:
    angle_ds = load_model_data.get_coastline_angle_kernel(
        compute=False,
        path_to_load="/g/data/ng72/ab4502/coastline_data/"+m+".nc",
        lat_slice=slice(-45.7,-6.9),
        lon_slice=slice(108,158.5)
    )
    angle_list.append(angle_ds)
```

```python
def load(model,diagnostic,coarsen=False,coarsen_lat=None,coarsen_lon=None,chunks={"time":10}):

    ds = xr.open_mfdataset("/g/data/ng72/ab4502/sea_breeze_detection/"+\
                      model+\
                      "/filters/filtered_mask_no_hourly_change_"+\
                      diagnostic+\
                      "_*.zarr/",
                 chunks={}, engine="zarr").chunk(chunks)

    if coarsen:
        ds = ds.groupby_bins("lat",coarsen_lat).max().groupby_bins('lon', coarsen_lon).max()
        ds = ds.rename({'lat_bins': 'lat', 'lon_bins': 'lon'})
        ds['lat'] = [(bins.left + bins.right) / 2 for bins in ds.lat.values]
        ds['lon'] = [(bins.left + bins.right) / 2 for bins in ds.lon.values]        
    
    return ds#.mask.resample({"time":"1D"}).max()#.sum("time").persist()

coarsen_lat = angle_list[-1].lat
coarsen_lon = angle_list[-1].lon

coarsen_lon = np.insert(coarsen_lon - (0.25/2), coarsen_lon.size ,  coarsen_lon.max()+(0.25/2))
coarsen_lat = np.insert(coarsen_lat - (0.25/2), coarsen_lat.size ,  coarsen_lat.max()+(0.25/2))
```

```python
f = load("aus2200_smooth_s4","F",coarsen=False,chunks={"time":-1,"lat":250,"lon":250}).mask.resample({"time":"1D"}).max().sum("time").persist()
```

```python
models = ["aus2200_smooth_s4","barra_c_smooth_s2","barra_r","era5"]
coarsen = [True,True,True,False]
F_list = []
for m,c in zip(models,coarsen):
    F_list.append(load(m,"F",filtered=True,coarsen=c,coarsen_lon=coarsen_lon,coarsen_lat=coarsen_lat))
```

```python
plt.figure(figsize=[18,12])

levs = np.arange(0,22,2)

ax=plt.subplot(2,2,1,projection=ccrs.PlateCarree())
F_list[0].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("F AUS2200 (smoothed sigma=4)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,2,projection=ccrs.PlateCarree())
F_list[1].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("F BARRA-C (smoothed sigma=2)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,3,projection=ccrs.PlateCarree())
F_list[2].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("F BARRA-R")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,4,projection=ccrs.PlateCarree())
F_list[3].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("F ERA5")
ax.coastlines(lw=0.2)
```

```python
models = ["aus2200_smooth_s4","barra_c_smooth_s2","barra_r","era5"]
coarsen = [True,True,True,False]
Fc_list = []
for m,c in zip(models,coarsen):
    Fc_list.append(load(m,"Fc",filtered=True,coarsen=c,coarsen_lon=coarsen_lon,coarsen_lat=coarsen_lat))
```

```python
plt.figure(figsize=[18,12])

levs = np.arange(0,22,2)

ax=plt.subplot(2,2,1,projection=ccrs.PlateCarree())
Fc_list[0].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fc AUS2200 (smoothed sigma=4)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,2,projection=ccrs.PlateCarree())
Fc_list[1].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fc BARRA-C (smoothed sigma=2)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,3,projection=ccrs.PlateCarree())
Fc_list[2].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fc BARRA-R")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,4,projection=ccrs.PlateCarree())
Fc_list[3].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fc ERA5")
ax.coastlines(lw=0.2)
```

```python
models = ["aus2200_smooth_s4","barra_c_smooth_s2","barra_r","era5"]
coarsen = [True,True,True,False]
fuzzy_list = []
diagnostic = ["fuzzy","fuzzy","fuzzy_mean","fuzzy_mean"]
for m,c,d in zip(models,coarsen,diagnostic):
    fuzzy_list.append(load(m,d,filtered=True,coarsen=c,coarsen_lon=coarsen_lon,coarsen_lat=coarsen_lat))
```

```python
plt.figure(figsize=[18,12])

levs = np.arange(0,22,2)

ax=plt.subplot(2,2,1,projection=ccrs.PlateCarree())
fuzzy_list[0].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fuzzy algorithm AUS2200 (smoothed sigma=4)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,2,projection=ccrs.PlateCarree())
fuzzy_list[1].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fuzzy algorithm (smoothed sigma=2)")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,3,projection=ccrs.PlateCarree())
fuzzy_list[2].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fuzzy algorithm BARRA-R")
ax.coastlines(lw=0.2)

ax=plt.subplot(2,2,4,projection=ccrs.PlateCarree())
fuzzy_list[3].plot(levels=levs,cmap="Blues",x="lon",y="lat")
plt.title("Fuzzy algorithm ERA5")
ax.coastlines(lw=0.2)
```
