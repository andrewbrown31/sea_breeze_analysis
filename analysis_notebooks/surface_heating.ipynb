```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
```

```python
era5 = xr.open_dataset("/g/data/ng72/ab4502/hourly_composites/era5/era5_ta_daily_max_2013_2016_2018.nc") - 273.15
barra_r = xr.open_dataset("/g/data/ng72/ab4502/hourly_composites/barra_r/barra_r_ta_daily_max_2013_2016_2018.nc") - 273.15
barra_c = xr.open_dataset("/g/data/ng72/ab4502/hourly_composites/barra_c/barra_c_ta_daily_max_2013_2016_2018.nc") - 273.15
aus2200 = xr.open_dataset("/g/data/ng72/ab4502/hourly_composites/aus2200/aus2200_ta_daily_max_2013_2016_2018.nc") - 273.15
```

```python
plt.figure(figsize=[16,12])

ax=plt.subplot(2,2,1,projection=ccrs.PlateCarree())
c = era5.t2m.plot(ax=ax,cmap="Reds",levels=np.linspace(10,38,9),extend="both",add_colorbar=False)
ax.coastlines()
ax.gridlines(ls=":",draw_labels=["bottom","left"])
plt.title("ERA5",size=18)

ax=plt.subplot(2,2,2,projection=ccrs.PlateCarree())
barra_r.tas.plot(ax=ax,cmap="Reds",levels=np.linspace(10,38,9),extend="both",add_colorbar=False)
ax.coastlines()
ax.gridlines(ls=":",draw_labels=["bottom","left"])
plt.title("BARRA-R",size=18)

ax=plt.subplot(2,2,3,projection=ccrs.PlateCarree())
barra_c.tas.plot(ax=ax,cmap="Reds",levels=np.linspace(10,38,9),extend="both",add_colorbar=False)
ax.coastlines()
ax.gridlines(ls=":",draw_labels=["bottom","left"])
plt.title("BARRA-C",size=18)

ax=plt.subplot(2,2,4,projection=ccrs.PlateCarree())
aus2200.ta.plot(ax=ax,cmap="Reds",levels=np.linspace(10,38,9),extend="both",add_colorbar=False)
ax.coastlines()
ax.gridlines(ls=":",draw_labels=["bottom","left"])
plt.title("AUS2200",size=18)

cb=plt.colorbar(c, cax=plt.axes([0.2,0.05,0.6,0.02]),orientation="horizontal")
cb.set_label("Average daily maximum temperature (deg C)")

plt.savefig("/g/data/ng72/ab4502/figs/sea_breeze/surface_ta_avg.jpeg",dpi=600,bbox_inches="tight")
```
