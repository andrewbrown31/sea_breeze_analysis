```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import iris
```

```python
pres = iris.load("/scratch/gb02/ab4502/cylc-run/u-dg768/share/cycle/20220227T0000Z/Lismore/d0198/RAL3P2/um/umnsaa_pb012",
              "air_pressure_at_sea_level")
pres_ds = xr.DataArray.from_iris(pres[0])
```

```python
u = iris.load("/scratch/gb02/ab4502/cylc-run/u-dg768/share/cycle/20220227T0000Z/Lismore/d0198/RAL3P2/um/umnsaa_pvera012",
              "x_wind")
u_ds = xr.DataArray.from_iris(u[0])

v = iris.load("/scratch/gb02/ab4502/cylc-run/u-dg768/share/cycle/20220227T0000Z/Lismore/d0198/RAL3P2/um/umnsaa_pvera012",
              "y_wind")
v_ds = xr.DataArray.from_iris(v[0])

wind_ds = xr.Dataset({"u":u_ds,"v":v_ds})
```

```python
step=20
p_levs = np.linspace(996,1016,11)*100
plt.figure(figsize=[14,14])

for time, pos in zip(["2022-02-27 15:00","2022-02-27 18:00","2022-02-27 21:00","2022-02-28 00:00"], [1,2,3,4]):
    ax = plt.subplot(2,2,pos,projection=ccrs.PlateCarree())
    xr.plot.contour(pres_ds.sel(time=time),colors=["k"],levels=p_levs,linewidths=[0.5])
    xr.plot.contourf(pres_ds.sel(time=time),cmap="Blues_r",alpha=0.9,levels=p_levs)
    ax.coastlines()
    ax.gridlines(draw_labels=["left","bottom"],ls=":")
    
    wind_ds.sel(time=time).isel(latitude=slice(0,-1,step),longitude=slice(0,-1,step)).plot.quiver("longitude","latitude","u","v",width=0.004,color="k")
```
